```python
# esta formula se realiza para obtener la diferencia entre un trimestre con respecto al anterior.

# Obtengo los valores de accesos por cada 100 hogares en un período anterior
accesos_anterior = df1_int_hog.groupby(['Provincia'])['Accesos por cada 100 hogares'].shift(-1)
# Calculo el aumento del 2% en el período anterior
aumento = accesos_anterior * 1.02
# Asigno los valores calculados al trimestre siguiente
df1_int_hog.loc[:, 'Accesos por cada 100 hogares trimestre anterior + 2%'] =round(aumento,2)
```

```python
#Remplazo los valores nulos por los valores en la fila de la columna 'Accesos por cada 100 hogares'
df1_int_hog['Accesos por cada 100 hogares trimestre anterior + 2%'] = df1_int_hog['Accesos por cada 100 hogares trimestre anterior + 2%'].fillna(df1_int_hog['Accesos por cada 100 hogares'])
```

```python
#Realizo una columna que me indique cuanto fue la diferencia entre ellos.
df1_int_hog['Diferencia'] = round(df1_int_hog['Accesos por cada 100 hogares trimestre anterior + 2%']-  df1_int_hog['Accesos por cada 100 hogares'],2)
```

```python
#Luego que me lo calcule en porcenteja.
df1_int_hog['Porcentaje de diferencia'] = round(df1_int_hog['Diferencia'] / df1_int_hog['Accesos por cada 100 hogares'] * 100,2)
```

```python
df1_int_hog.tail()
```

Realizo un grafico de dispersion para ver como se distribuye el acceso en las provincias

```python
# Creo la figura y el eje
fig, ax = plt.subplots(figsize=(10, 5))

# Realizo el gráfico de dispersión
ax.scatter(df1_int_hog["Accesos por cada 100 hogares"], df1_int_hog["Provincia"])

# Agrego etiquetas y título
ax.set_xlabel("Accesos por cada 100 hogares")
ax.set_ylabel("Provincia")
ax.set_title("Gráfico de dispersión")
plt.show()
```

Realizo un Boxplot para verificar posibles outliers

```python
sns.boxplot(data=df1_int_hog, y='Provincia', x='Accesos por cada 100 hogares')
```

Realizo un histograma para ver de mayor a menor cuales son las provincias con mayor concentracion de accesos por hogares.

```python
provincia_group = df1_int_hog.groupby('Provincia').mean().sort_values('Accesos por cada 100 hogares', ascending=False)
provincia_group.plot(kind='bar', y='Accesos por cada 100 hogares', figsize=(10, 5))
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Promedio de accesos por provincia')
plt.show()
```

Aca lo realizo para ver como se distribuyo en cada trimestre.

```python
fig, ax = plt.subplots(figsize=(12, 6))
# Agrupo por año y trimestre y calcular la media de accesos por cada 100 hogares
ano_trimestre_group = df1_int_hog.groupby(['Año', 'Trimestre']).mean(numeric_only=True)
                                                     
# Genero la gráfica
plt.bar(np.arange(len(ano_trimestre_group)), ano_trimestre_group['Accesos por cada 100 hogares'])
# Configuro los ejes y el título de la gráfica
plt.xticks(np.arange(len(ano_trimestre_group)), ano_trimestre_group.index, rotation=90)
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Promedio de accesos por año y trimestre')
plt.show()
```
